## α) Διηλεκτρική Σταθερά: Πραγματικό και Φανταστικό Μέρος

Στο παρακάτω κελί υπολογίζουμε και απεικονίζουμε το **πραγματικό (Re)** και το **φανταστικό (Im)** μέρος της διηλεκτρικής σταθεράς \(\epsilon(\omega)/\epsilon_0\) για ένα απλό flat-top μοντέλο.

### Δίνεται

Το φανταστικό μέρος δίνεται από:

$$
\Im\left(\varepsilon / \varepsilon_0\right) = \lambda \left[ \theta(\omega - \omega_1) - \theta(\omega - \omega_2) \right], 
\quad \omega_2 > \omega_1 > 0
$$

δηλαδή, παίρνει σταθερή τιμή $\lambda$ μόνο μεταξύ $\omega_1$ και $\omega_2$ και είναι μηδέν έξω από αυτή την περιοχή. Η συνάρτηση $\theta$ είναι η συνάρτηση Heaviside.

### Υπολογισμός του πραγματικού μέρους

Χρησιμοποιώντας τις **σχέσεις Kramers-Kronig**, το πραγματικό μέρος υπολογίζεται ως:

$$
\Re\left(\epsilon(\omega) / \epsilon_0\right) = 1 + \frac{\lambda}{\pi} 
\left[ \ln\left|\omega_2^2 - \omega^2\right| - \ln\left|\omega_1^2 - \omega^2\right| \right]
$$

Δηλαδή, μέσω του ολοκληρώματος principal value που δίνει η Kramers-Kronig σχέση, προκύπτει η πραγματική συνιστώσα της διηλεκτρικής σταθεράς.

### Τι απεικονίζουμε

- Το **πραγματικό μέρος** με **συνεχή μαύρη γραμμή**.
- Το **φανταστικό μέρος** με **διακεκομμένη μαύρη γραμμή**.
- Οριζόντιος άξονας: συχνότητα $\omega$
- Κατακόρυφος άξονας: $\epsilon(\omega)/\epsilon_0$

Με αυτόν τον τρόπο μπορούμε να δούμε:
- Πού εμφανίζεται **anomalous dispersion** λόγω flat-top Im.
- Πώς οι περιοχές απώλειας στο Im επηρεάζουν το Re.

```python
import numpy as np
import matplotlib.pyplot as plt

# --- Παράμετροι ---
lambda_ = 1.0     # συντελεστής λ
omega1 = 1.0      # χαμηλή συχνότητα
omega2 = 3.0      # υψηλή συχνότητα

# --- Συχνότητες ---
omega = np.linspace(0, 5, 1000)

# --- Πραγματικό μέρος (Re) ---
Re_eps = 1 + (lambda_/np.pi)*(np.log(np.abs(omega2**2 - omega**2)) - np.log(np.abs(omega1**2 - omega**2)))

# --- Φανταστικό μέρος (Im) ---
# Flat top between omega1 and omega2
Im_eps = np.zeros_like(omega)
Im_eps[(omega >= omega1) & (omega <= omega2)] = lambda_

# --- Plotting ---
plt.figure(figsize=(10,5))

# Πραγματικό μέρος
plt.plot(omega, Re_eps, 'k', lw=2, label=r'$\mathrm{Re}(\epsilon/\epsilon_0)$')

# Φανταστικό μέρος
plt.plot(omega, Im_eps, 'k--', lw=2, label=r'$\mathrm{Im}(\epsilon/\epsilon_0)$')

# Στυλ "paper style"
plt.xlabel(r'$\omega$', fontsize=14)
plt.ylabel(r'$\epsilon / \epsilon_0$', fontsize=14)
plt.title('Διηλεκτρική Σταθερά', fontsize=16)
plt.legend(frameon=False, fontsize=12)
plt.grid(True, linestyle=':', alpha=0.5)
plt.tight_layout()
plt.savefig("a.png", dpi=300)
plt.show()
```

## β) Διηλεκτρική Σταθερά

Στο υποερώτημα (β) έχουμε φανταστικό μέρος της διηλεκτρικής σταθεράς:

$$
\Im\left(\varepsilon / \varepsilon_0\right) = \frac{\lambda \gamma \omega}{(\omega_0^2 - \omega^2)^2 + \gamma^2 \omega^2}
$$

δηλαδή πρόκειται για **κλασικό Lorentzian** που αντιπροσωπεύει ένα **resonance** γύρω από $\omega_0$ με πλάτος $\gamma$.

### Υπολογισμός του πραγματικού μέρους

Χρησιμοποιώντας τις **σχέσεις Kramers-Kronig**, καταλήγουμε στο:

$$
\Re\left(\epsilon(\omega) / \epsilon_0\right) = 1 + \lambda \frac{\omega_0^2 - \omega^2}{(\omega_0^2 - \omega^2)^2 + \gamma^2 \omega^2}
$$

### Τι απεικονίζουμε

- Το **πραγματικό μέρος** $\Re(\epsilon/\epsilon_0)$ με συνεχή γραμμή.
- Το **φανταστικό μέρος** $\Im(\epsilon/\epsilon_0)$ με διακεκομμένη γραμμή.
- Οριζόντιος άξονας: συχνότητα $\omega$
- Κατακόρυφος άξονας: $\epsilon(\omega)/\epsilon_0$

```python
# --- Παράμετροι ---
lambda_ = 1.0
omega0 = 2.0
gamma = 0.2

# --- Συχνότητες ---
omega = np.linspace(0, 4, 1000)

# --- Φανταστικό μέρος (Im) ---
Im_eps = (lambda_ * gamma * omega) / ((omega0**2 - omega**2)**2 + (gamma**2 * omega**2))

# --- Πραγματικό μέρος (Re) ---
Re_eps = 1 + lambda_ * (omega0**2 - omega**2) / ((omega0**2 - omega**2)**2 + gamma**2 * omega**2)

# --- Plotting ---
plt.figure(figsize=(10,5))
plt.plot(omega, Re_eps, 'k', lw=2, label=r'$\mathrm{Re}(\epsilon/\epsilon_0)$')
plt.plot(omega, Im_eps, 'k--', lw=2, label=r'$\mathrm{Im}(\epsilon/\epsilon_0)$')

plt.xlabel(r'$\omega$', fontsize=14)
plt.ylabel(r'$\epsilon / \epsilon_0$', fontsize=14)
plt.title('Διηλεκτρική Σταθερά: Lorentzian Resonance', fontsize=16)
plt.legend(frameon=False, fontsize=12)
plt.grid(True, linestyle=':', alpha=0.5)
plt.tight_layout()
plt.savefig("b.png", dpi=300)
plt.show()
```
