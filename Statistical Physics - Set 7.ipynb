```python
import numpy as np
import matplotlib.pyplot as plt

# Παράμετροι
N = 1
k = 1
eps = 1

T = np.linspace(0.05, 5, 1000)

# εβ = 1/T επειδή ε/k = 1
x = 1 / T

U = -N * eps * np.tanh(x)

S = N * k * (
    np.log(2 * np.cosh(x))
    - x * np.tanh(x)
)

Cv = N * k * x**2 / np.cosh(x)**2

# Δημιουργία figure 1x3
fig, ax = plt.subplots(1, 3, figsize=(14, 4))

# U(T)
ax[0].plot(T, U, color="black")
ax[0].set_title(r'$U(T)$')
ax[0].set_xlabel(r'$T$')
ax[0].set_ylabel(r'$U$')
ax[0].grid(True)

# S(T)
ax[1].plot(T, S, color="black")
ax[1].set_title(r'$S(T)$')
ax[1].set_xlabel(r'$T$')
ax[1].set_ylabel(r'$S$')
ax[1].grid(True)

# Cv(T)
ax[2].plot(T, Cv, color="black")
ax[2].set_title(r'$C_V(T)$')
ax[2].set_xlabel(r'$T$')
ax[2].set_ylabel(r'$C_V$')
ax[2].grid(True)

plt.tight_layout()
plt.savefig("1.png",dpi=300)
plt.show()
```
