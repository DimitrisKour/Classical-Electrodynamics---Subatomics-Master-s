### Example 5.5 (c):

For the case $\beta=\pi$ of a plane conductor with a half cylinder of radius $\alpha$ on it held at zero potential, we found that far from the plane, the two limiting induced surface-charge densities:

$$\sigma(\alpha,\phi)=\sigma_0\sin\phi,\quad\sigma_0\equiv-A_1\frac{2\epsilon_0}{\alpha}$$

$$\sigma(\rho,0)=\sigma(\rho,\beta)=\frac{\sigma_0}{2}\left[1-\left(\frac{\rho}{\alpha}\right)^{-2}\right]$$

Below, we sketch the charge density on and in the neighborhood of the half-cylinder.

```python
rho_min = 1.0
rho_max = 4.0

def sigma_phi_over_sigma0(phi):
    return np.sin(phi)

def sigma_rho_over_sigma0(rho):
    return 0.5 * (1 - (rho)**(-2))

phi = np.linspace(0, np.pi, 400)
sigma_phi = sigma_phi_over_sigma0(phi)

rho = np.linspace(rho_min, rho_max, 400)
sigma_rho = sigma_rho_over_sigma0(rho)
```

```python
fig, ax = plt.subplots(figsize=(7, 5.5))

ax.plot(-rho, sigma_rho, color="black")
ax.plot((phi - np.pi/2)/(np.pi/2), sigma_phi, color="black")
ax.plot(+rho, sigma_rho, color="black")


ax.grid(True, linestyle="--", alpha=0.35)
ax.set_xlim(-rho_max, rho_max)
ax.set_ylim(0, 1.05)
ax.set_xlabel("")  
fs = 16
ax.text(0.15, -0.15, r"$\rho/a$", transform=ax.transAxes,
        ha="left", va="center", fontsize=fs)
ax.text(0.5, -0.15, r"$\dfrac{2\phi-\pi}{\pi}$", transform=ax.transAxes,
        ha="center", va="center", fontsize=fs)
ax.text(0.85, -0.15, r"$\rho/a$", transform=ax.transAxes,
        ha="right", va="center", fontsize=fs)

ax.set_ylabel(r"$\sigma / \sigma_0$")

plt.tight_layout()
plt.savefig("5.5.png", dpi=300)
plt.show()
```
